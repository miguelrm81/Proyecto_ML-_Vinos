
- winery: Winery name
- wine: Name of the wine
- year: Year in which the grapes were harvested
- rating: Average rating given to the wine by the users [from 1-5]
- num_reviews: Number of users that reviewed the wine
- country: Country of origin [Spain]
- region: Region of the wine
- price: Price in euros [€]
- type: Wine variety
- body: Body score, defined as the richness and weight of the wine in your mouth [from 1-5]
- acidity: Acidity score, defined as wine's “pucker” or tartness; it's what makes a wine refreshing and your tongue salivate and want   another sip [from 1-5]

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = df[['rating', 'num_reviews', 'price', 'acidity', 'body']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriz de Correlación')
plt.show()
```

```python

df2['rating_normalizado'] = df2['rating'] / df2['rating'].max()

df2['rating ponderado'] = df2['rating_normalizado'] * df2['num_reviews']


```

```python
df2.to_csv('wines_SPA_proce.csv', index=False)
```

```python
df2['valoracion'] = 0


for index, row in df2.iterrows():
    if row['rating ponderado'] > 500:
        df2.loc[index, 'valoracion'] = 3
    elif 350 <= row['rating ponderado'] <= 500:
        df2.loc[index, 'valoracion'] = 2
    else :
        df2.loc[index, 'valoracion'] = 1
   
```

```python
df2['valoracion'].value_counts()
```

```python
df2['valoracion'].value_counts(normalize= True).plot(kind = 'bar')
plt.xlabel("Clase de valoracion")
plt.ylabel("Frequenccia")
plt.title("Balance de las valoraciones");
```
